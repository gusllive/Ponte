```python
import numpy as np
import matplotlib.pyplot as plt

# Definir as variáveis
Ca0 = 0.1  # Concentração inicial mol/L
k = 20    # Constante de taxa Mol/L min
n = 2.0    # Ordem da reação

# Gerar uma faixa de valores de Ca
Ca_values = np.linspace(0.01, Ca0, 100)

# Calcular Tau em função de Ca
Tau_values = (Ca0 - Ca_values) / (k * Ca_values**n)

# Plotar a relação
plt.plot( Tau_values,Ca_values, label= 'Ca vs Tau')
plt.ylabel('Ca')
plt.xlabel('Tau')
plt.title('Plot of Ca vs Tau')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Find the indices where Ca_values is greater than 0.49
indices = np.where(Ca_values > 0.05)

# Print the corresponding Tau_values
print(Tau_values[indices])
```
